```python
import numpy as np
import pickle
import os
import matplotlib.pyplot as plt
import matplotlib
```

```python
folders = ["C:/Users/Ivo/Documents/qiskit_codes/data/sherbrooke/46/power_broadening (narrowing)/ipN_pulses/2024-11-27",
           "C:/Users/Ivo/Documents/qiskit_codes/data/sherbrooke/46/power_broadening (narrowing)/ipN_pulses/2024-11-29"]
times = []
for folder in folders:
    times_folder = [folder + '/' + f for f in os.listdir(folder)]
    times.extend(times_folder)
print(times)
```

```python
l, p, x0 = 396.5566394925376,0.3037880174467387,0.0010160433792621
T = 800 * 0.5e-9
tr_probs, dets, amps = [], [], []
for t in times: 
    files = os.listdir(t)
    f1 = [f for f in files if f.endswith("tr_prob.pkl")][0]
    f2 = [f for f in files if f.endswith("detunings.pkl")][0]
    f3 = [f for f in files if f.endswith("areas.pkl")][0]
    with open(os.path.join(t, f1), "rb") as f1:
        tr_probs.append(pickle.load(f1))    
    with open(os.path.join(t, f2), "rb") as f2:
        dets.append(pickle.load(f2) / 1e6)   
    with open(os.path.join(t, f3), "rb") as f3:
        amps.append(l * (1 - np.exp(- p * (pickle.load(f3) - x0))) / (1e6 * T))
```

```python
fontsize = 18
fig = plt.figure(figsize=(14,12), layout="constrained")
gs = fig.add_gridspec(2, 3, width_ratios=[1, 1, .1])
cmap = matplotlib.colormaps["cividis"]
for i in range(len(times)):
    ax = fig.add_subplot(gs[int(i / 2), int(i % 2)])
    c = ax.pcolormesh(dets[i], amps[i], tr_probs[i], cmap=cmap)
    xticks = np.linspace(dets[i][0], dets[i][-1], 7)
    xticklabels = np.round(np.linspace(dets[i][0], dets[i][-1], 7),2)
    yticks = np.linspace(amps[i][0], amps[i][-1], 6)
    yticklabels = np.round(np.linspace(amps[i][0], amps[i][-1], 6),2)
    if int(i/2) == 1:
        ax.set_xticks(xticks)
        ax.set_xticklabels(xticklabels, fontsize=fontsize)
        ax.set_xlabel("Detuning (MHz)", fontsize=fontsize)
    else:
        ax.set_xticks([])
    if int(i%2) == 0:
        ax.set_yticks(yticks)
        ax.set_yticklabels(yticklabels, fontsize=fontsize)
        ax.set_ylabel("Rabi freq. (arb. units)", fontsize=fontsize)

    ax.set_yticks(yticks)
    ax.set_yticklabels(np.round(yticklabels).astype(int), fontsize=fontsize)
    c.set_clim(0,1)
cax = fig.add_subplot(gs[:, 2])
cbar = fig.colorbar(c, cax=cax) 
cbar.ax.tick_params(labelsize=fontsize)
colorbar_ticks = np.round(np.arange(0, 1.01, 0.2),1)
cbar.set_ticks(colorbar_ticks)
```

```python
# Define the function
def func(t, N, T=1):
    return ((t - T/2) / (T/2)) ** (2 * N)

```

```python
# Generate t values in the interval [0, 1]
t = np.linspace(0, 1, 400)

# Create a 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# List of N values
N_values = [0, 1, 2, 3]

# Titles for the subplots
titles = ["N=0", "N=1", "N=2", "N=3"]

fontsize = 15

# Plot the functions
for i, ax in enumerate(axs.flat):
    N = N_values[i]
    y = func(t, N)
    ax.plot(t, y, label=f'N={N}', linewidth=2.75)
    ax.set_xlim(-0.1, 1.1)
    ax.set_ylim(-0.1, 1.15)
    ax.fill_between(t, y, color='skyblue', alpha=0.4)  # Fill the area under the curve    
    ax.grid(True, which='major')  # Enable major grid lines
    ax.grid(True, which='minor', alpha=0.2)  # Enable minor grid lines with alpha=0.2
    ax.minorticks_on()  # Enable minor ticks
    ax.axvline(x=0, ymin=1/12.5, ymax=11/12.5, color='#1f77b4', linestyle='-', linewidth=2.75)  # Vertical line at start
    ax.axvline(x=1, ymin=1/12.5, ymax=11/12.5, color='#1f77b4', linestyle='-', linewidth=2.75)  # Vertical line at end
    yticks = np.round(np.linspace(0, 1, 6), 1)
    xticks = np.round(np.linspace(0, 1, 6), 1)
    if i in [0, 2]:
        ax.set_ylabel("Rabi freq. $\Omega$ (arb. units)", fontsize=fontsize)
        ax.set_yticks(yticks)
        ax.set_yticklabels(yticks, fontsize=fontsize)
    else:
        ax.set_yticklabels([])
    if i in [2, 3]:
        ax.set_xlabel("Time t (in units T)", fontsize=fontsize)
        ax.set_xticks(xticks)
        ax.set_xticklabels(xticks, fontsize=fontsize)
    else:
        ax.set_xticklabels([])

    ax.legend(fontsize=fontsize)
# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Define the function
def fcq(t, beta, T=1):
    return (1 + beta * (((t - T/2) / (T/2)) ** 2 - 1))
```

```python
# Generate t values in the interval [0, 1]
t = np.linspace(0, 1, 400)

# Create a 2x2 grid of plots
fig, axs = plt.subplots(3, 2, figsize=(10, 15))

# List of N values
beta_values = [-1, 0, 0.25, 0.5, 0.75, 1]

# Titles for the subplots
titles = [f"$\\beta$={b}" for b in beta_values]

# Plot the functions
for i, ax in enumerate(axs.flat):
    beta = beta_values[i]
    y = fcq(t, beta)
    ax.plot(t, y, label=f'$\\beta$={beta}')
    ax.set_xlim(-0.1, 1.1)
    if i == 0:
        ax.set_ylim(-0.1, 2.1)
    else:
        ax.set_ylim(-0.1, 1.3)
    ax.grid(True, which='major')  # Enable major grid lines
    ax.grid(True, which='minor', alpha=0.2)  # Enable minor grid lines with alpha=0.2
    ax.minorticks_on()  # Enable minor ticks
    ax.axvline(x=0, ymin=1/14, ymax=11/14, color='#1f77b4', linestyle='-', linewidth=1)  # Vertical line at start
    ax.axvline(x=1, ymin=1/14, ymax=11/14, color='#1f77b4', linestyle='-', linewidth=1)  # Vertical line at end
    if i in [0, 2, 4]:
        ax.set_ylabel("Rabi freq. $\\Omega$ (arb. units)")
    if i in [4, 5]:
        ax.set_xlabel("Time t (in units T)")
    else:
        ax.set_xticklabels([])
    ax.legend()
# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
