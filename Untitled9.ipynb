```python
def make_all_dirs(path):
    folders = path.split("/")
    for i in range(2, len(folders) + 1):
        folder = "/".join(folders[:i])
        if not os.path.isdir(folder):
            os.mkdir(folder)

def get_calib_params(
    backend, qubit,
    pulse_type, 
    sigma, duration,
    remove_bg
):
    file_dir = os.getcwd()
    # file_dir = os.path.split(file_dir)[0]
    calib_dir = os.path.join(file_dir, "calibrations", backend, str(qubit))
    params_file = os.path.join(calib_dir, "actual_params.csv")
    print(params_file)
    if os.path.isfile(params_file):
        param_df = pd.read_csv(params_file)
    df = param_df[param_df.apply(
            lambda row: row["pulse_type"] == pulse_type and \
                row["duration"] == duration and \
                row["sigma"] == sigma and \
                row["rb"] == remove_bg, axis=1)]
    # print(df)
    idx = 0 
    if df.shape[0] > 1:
        idx = input("More than one identical calibrations found! "
                    "Which do you want to use? ")
    elif df.shape[0] < 1:
        raise ValueError("No entry found!")
    ser = df.iloc[idx]
    l = ser.at["l"]
    p = ser.at["p"]
    x0 = ser.at["x0"]

    return l, p, x0

def initialize_backend(backend):
    backend_full_name = "ibm_" + backend 
    GHz = 1.0e9 # Gigahertz
    MHz = 1.0e6 # Megahertz
    us = 1.0e-6 # Microseconds
    ns = 1.0e-9 # Nanoseconds
    qubit = 0
    mem_slot = 0

    drive_chan = pulse.DriveChannel(qubit)
    # meas_chan = pulse.MeasureChannel(qubit)
    # acq_chan = pulse.AcquireChannel(qubit)
    
    backend_name = backend
    # provider = IBMQ.load_account()
    ibm_backend = QiskitRuntimeService(channel="ibm_quantum").backend(backend_full_name)
    print(f"Using {backend_name} backend.")
    pm = generate_preset_pass_manager(optimization_level=0, backend=ibm_backend)
    backend_defaults = ibm_backend.defaults()
    backend_config = ibm_backend.configuration()

    center_frequency_Hz = backend_defaults.qubit_freq_est[qubit]# 4962284031.287086 Hz
    num_qubits = backend_config.n_qubits

    q_freq = [backend_defaults.qubit_freq_est[q] for q in range(num_qubits)]
    dt = backend_config.dt

    return ibm_backend, pm, drive_chan, num_qubits, q_freq

def fit_function(
    x_values, 
    y_values, 
    function, 
    init_params,
    bounds
):
    fitparams, conv = curve_fit(
        function, 
        x_values, 
        y_values, 
        init_params, 
        maxfev=100000, 
        bounds=bounds
    )
    y_fit = function(x_values, *fitparams)
    
    return fitparams, y_fit

def linear_func(x, a, b):
    return a * x + b
```

```python
def add_circ(ibm_backend, drive_chan, q_freq, amp, duration, sigma, qubit=0):
    with pulse.build(name="rabi_osc") as sched:
        dur_dt = duration
        
        if pulse_type == "sq" or "sin" in pulse_type:
            pulse_played = pulse_dict[pulse_type][remove_bg](
                duration=dur_dt,
                amp=amp,
                name=pulse_type
            )
        elif pulse_type == "gauss":
            pulse_played = pulse_dict[pulse_type][remove_bg](
                duration=dur_dt,
                amp=amp,
                name=pulse_type,
                sigma=sigma
            )
        elif pulse_type in ["lor", "lor2", "lor3"]:
            pulse_played = pulse_dict[pulse_type][remove_bg](
                duration=dur_dt,
                amp=amp,
                name=pulse_type,
                sigma=sigma,
            )
        elif pulse_type == "drag":
            pulse_played = pulse_dict[pulse_type][remove_bg](
                duration=dur_dt,
                amp=amp,
                beta=1,
                name=pulse_type,
                sigma=sigma,
            )
        elif pulse_type == "ipN":
            pulse_played = pulse_dict[pulse_type][remove_bg](
                duration=dur_dt,
                amp=amp,
                N=N,
                name=pulse_type
            )
        elif pulse_type == "fcq":
            pulse_played = pulse_dict[pulse_type][remove_bg](
                duration=dur_dt,
                amp=amp,
                beta=beta,
                name=pulse_type
            )
        else:
            pulse_played = pulse_dict[pulse_type][remove_bg](
                duration=dur_dt,
                amp=amp,
                name=pulse_type,
                sigma=sigma,
            )
        pulse.set_frequency(q_freq, drive_chan)
        pulse.play(pulse_played, drive_chan)
    pi_gate = Gate("rabi", 1, [])
    base_circ = QuantumCircuit(127, 1)
    base_circ.append(pi_gate, [qubit])
    base_circ.add_calibration(pi_gate, (qubit,), sched, [])
    base_circ.measure(qubit, 0)
    return base_circ
```

```python
pulse_type, qubit, duration, sigma = "sq", 46, 96, 96
num_T, T_min, T_max, num_amp, amp_min = 8, 160, 160, 100, 0.01
remove_bg, max_experiments_per_job = 1, 100
num_shots, num_exp, backend = 256, 100, "sherbrooke"
N, beta = 1, 0
save = 1
```

```python
l, p, x0 = get_calib_params(
    backend, qubit,
    pulse_type, 
    sigma, duration,
    remove_bg
)

three_pi_amp = -np.log(1 - 3 * np.pi / l) / p + x0 

ibm_backend, pm, drive_chan, num_qubits, q_freq = initialize_backend(backend)
Ts = np.logspace(np.log10(T_min), np.log10(T_max), num_T)
Ts = np.array([int(round(t / 16) * 16) for t in Ts])
amplitude_factors = Ts / Ts[0]

amplitudes = np.linspace(amp_min, three_pi_amp, num_amp)
amplitudes = amplitudes[None] / amplitude_factors[:, None]

circs = []
for idx, duration in enumerate(Ts):
    sigma = 0.1 * duration if pulse_type != "sq" else duration
    circs_dur = []
    for a in amplitudes[idx]:
        circs_dur.append(add_circ(ibm_backend, drive_chan, q_freq, a, duration, sigma, qubit=qubit))
    circs.append(circs_dur)

values, job_ids = [], []
for circs_dur, duration in zip(circs, Ts):
    print(len(circs_dur))
    print(circs_dur[0].draw)
    pm_circs = pm.run(circs_dur)
    res = run_jobs(pm_circs, ibm_backend, duration, num_shots_per_exp=num_shots_per_exp, pm=pm)
    values.append(res[0])
    job_ids.append(res[1])
```
