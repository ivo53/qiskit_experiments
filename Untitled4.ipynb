```python
import pickle
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt
```

```python
times = {
    "sq": ["2023-12-08", "175156"],
    "sin": ["2023-12-08", "130503"],
}
backend_name = "kyoto"
save_fig = 0
save_fig = 1
file_dir = os.getcwd()

def data_folder(date, time, pulse_type):
    return os.path.join(
        file_dir,
        "data",
        f"{backend_name}",
        "power_broadening (narrowing)",
        f"{pulse_type}_pulses",
        date,
        time
    ).replace("\\", "/")

l, p, x0 = 1.25116799e+02, 3.01445760e-01, 8.23116906e-03
T = 192 * 5e-10
tr_prob, amp, det = [], [], []
for k, t in times.items():
    files = os.listdir(data_folder(t[0], t[1], k))
    with open(os.path.join(data_folder(t[0], t[1], k), files[2]).replace("\\","/"), 'rb') as f1:
        tr_prob.append(pickle.load(f1))
    with open(os.path.join(data_folder(t[0], t[1], k), files[0]), 'rb') as f2:
        amp.append(l * (1 - np.exp(-p * (pickle.load(f2) - x0))) / (1e6 * (2*T/np.pi)))
    with open(os.path.join(data_folder(t[0], t[1], k), files[1]), 'rb') as f3:
        det.append(pickle.load(f3) * 2 * np.pi / 1e6)

# with open("C:/Users/Ivo/Documents/qiskit_experiments/sine.pkl", "rb") as f1:
#     data = pickle.load(f1)
# interval_det, interval_amp = [100, 100], [100, 150]
# lim_det, lim_amp = [230, 230], [300, 500]#490]
# det.append(np.arange(-2.3e2, 2.3001e2, 0.05e2))
# amp.append(np.arange(0, 5.001e2, 2.5))
# # tr_prob.append(data[:, 2].reshape(len(det[2]), len(amp[2])).T)
# tr_prob.append(data)

# # data_sq = pd.read_csv("C:/Users/Ivo/Documents/qiskit_experiments/sq.csv", header=None).to_numpy()
# with open("C:/Users/Ivo/Documents/qiskit_experiments/sq.pkl", "rb") as f2:
#     data_sq = pickle.load(f2)
# det.insert(1, np.arange(-2.3e2, 2.3001e2, 0.025e2))
# amp.insert(1, np.arange(0, 3.20001e2, 1))
# tr_prob.insert(1, data_sq)#[:, 2].reshape(len(det[1]), len(amp[1])).T)
```

```python
with open("./pulse_optimisation/result_drag_2024-08-19_213423.pkl", "rb") as f1:
    result_drag = pickle.load(f1)
with open("./pulse_optimisation/result_rect_2024-08-19_213423.pkl", "rb") as f2:
    result_rect = pickle.load(f2)
with open("./pulse_optimisation/result_sine_2024-08-19_213423.pkl", "rb") as f3:
    result_sine = pickle.load(f3)
```

```python
A_drag_max, A_rect_max, A_sine_max = (1.3036867304289093, 0.3272492347489368, 0.5140418958900708)
d_min, d_max, d_num = (-0.23, 0.23, 101)
A_min, A_num = (0, 100)
d_range = np.linspace(d_min, d_max, d_num) * 1e3

A_drag_range = np.linspace(A_min, A_drag_max, A_num) * 1e3
A_rect_range = np.linspace(A_min, A_rect_max, A_num) * 1e3
A_sine_range = np.linspace(A_min, A_sine_max, A_num) * 1e3
tr_probs_drag = np.array([[np.abs(r[1,1]) for r in res] for res in result_drag])[::-1]
tr_probs_rect = np.array([[np.abs(r[1,1]) for r in res] for res in result_rect])[::-1]
tr_probs_sine = np.array([[np.abs(r[1,1]) for r in res] for res in result_sine])[::-1]

tr_prob.extend([tr_probs_rect.T, tr_probs_sine.T])
amp.extend([A_rect_range, A_sine_range])
det.extend([d_range, d_range])
```

```python
from datetime import datetime
dt_now = datetime.now()
```

```python
def leakage(result_optim):
    leakage_values = []
    for res in result_optim:
        leakage_line = []
        for r in res:
            leakage = 0
            for d in range(2, 8):
                leakage += np.abs(r[d,d])
            leakage_line.append(leakage)
        leakage_values.append(leakage_line)
    return np.array(leakage_values)
    
leakage_drag = leakage(result_drag)[::-1]
leakage_rect = leakage(result_rect)[::-1]
leakage_sine = leakage(result_sine)[::-1]

tr_prob.extend([leakage_rect.T, leakage_sine.T])
amp.extend([A_rect_range, A_sine_range])
det.extend([d_range, d_range])
```

```python
interval_det, interval_amp = [100, 100], [100, 150]
lim_det, lim_amp = [230, 230], [300, 500]
```

```python
fig = plt.figure(figsize=(23, 14), layout="constrained")
gs = fig.add_gridspec(2, 5, width_ratios=[1, 1, 0.08, 1, 0.08])
cmap = plt.get_cmap("cividis")  # Choose a colormap
ims = []
for i in range(2):
    for j in range(3):
        a = amp[2 * j + i]
        d = det[2 * j + i]
        tr = tr_prob[2 * j + i]
        if j == 2:
            ax = fig.add_subplot(gs[i, 3])
        if j < 2:
            ax = fig.add_subplot(gs[i, j])
            ims.append(ax.imshow(tr, cmap=cmap, aspect="auto", origin="lower", vmin=0, vmax=1))
        elif i == 0:
            ims.append(ax.imshow(tr, cmap=cmap, aspect="auto", origin="lower", vmin=0, vmax=1e-2))
            # ax.set_title("Leakage")
            # ax.legend(title="Leakage in units $10^{-2}$")
        elif i == 1:
            ims.append(ax.imshow(tr, cmap=cmap, aspect="auto", origin="lower", vmin=0, vmax=1e-5))
            # ax.legend(title="Leakage in units $10^{-5}$")
            
        max_amp = interval_amp[i] * np.floor(lim_amp[i] / interval_amp[i])
        max_det = interval_det[i] * np.floor(lim_det[i] / interval_det[i])
                
        ax.set_xlim(
            (
                len(tr[0])-int(np.round(lim_det[i] / d[-1] * (len(tr[0]) / 2 - 1) + len(tr[0]) / 2)), 
                int(np.round(lim_det[i] / d[-1] * (len(tr[0]) / 2 - 1) + len(tr[0]) / 2))
            )
        )
        ax.set_ylim(
            (
                0,
                int(np.round(lim_amp[i] / a[-1] * (len(tr) - 1)))
            )
        )

        xticks = np.linspace(
            len(tr[0])-int(np.round(max_det / d[-1] * (len(tr[0]) / 2 - 1) + len(tr[0]) / 2)), 
            int(np.round(max_det / d[-1] * (len(tr[0]) / 2 - 1) + len(tr[0]) / 2)), 
            int(2 * max_det / interval_det[i] + 1)
        )
        yticks = np.linspace(
            0, 
            int(np.round(max_amp / a[-1] * (len(tr) - 1))), 
            int(max_amp / interval_amp[i] + 1)
        )
        ax.set_xticks(xticks)
        ax.set_yticks(yticks)
        if j == 0:
            ax.set_yticklabels([int(tick) for tick in np.linspace(0, max_amp, int(max_amp / interval_amp[i] + 1)).round(0)], fontsize=24)
            pulse_type = "Rectangular" if i == 0 else "Sinusoidal"
            ax.set_ylabel(f'{pulse_type}\nAmplitude (MHz)', fontsize=28)
        else:
            ax.set_yticklabels([])
        if i == 1:
            ax.set_xlabel('Detuning (MHz)', fontsize=28)
            ax.set_xticklabels([int(tick) for tick in np.linspace(-max_det, max_det, int(2 * max_det / interval_det[i] + 1)).round(0)], fontsize=24)
        else:
            if j == 0:
                title = "Observed\nTransition Probability"
            if j == 1:
                title = "Numerical\nTransition Probability"
            if j == 2:
                title = "Numerical\nLeakage"
            ax.set_title(title, fontsize=24)
            ax.set_xticklabels([])
# Create a separate subplot for the color bar
cax = fig.add_subplot(gs[:, 2])
cbar = fig.colorbar(ims[0], cax=cax)
# cbar.set_label('Transition probability', fontsize=15)
cbar.ax.tick_params(labelsize=28)

# Create a separate subplot for the color bar
cax2 = fig.add_subplot(gs[0, 4])
cbar2 = fig.colorbar(ims[2], cax=cax2)
# cbar.set_label('Transition probability', fontsize=15)
cbar2.ax.tick_params(labelsize=28)

# Create a separate subplot for the color bar
cax3 = fig.add_subplot(gs[1, 4])
cbar3 = fig.colorbar(ims[5], cax=cax3)
# cbar.set_label('Transition probability', fontsize=15)
cbar3.ax.tick_params(labelsize=28)
cbar3.ax.yaxis.get_offset_text().set_fontsize(18)

time = dt_now.strftime("%H%M%S")
date = dt_now.strftime("%Y-%m-%d")
plt.savefig(f"./paper_ready_plots/finite_pulses/f4_{date}_{time}.pdf")

# plt.show()
```
