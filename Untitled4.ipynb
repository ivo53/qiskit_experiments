```python
import pickle
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt
```

```python
with open("./pulse_optimisation/result_drag.pkl", "rb") as f1:
    result_drag = pickle.load(f1)
with open("./pulse_optimisation/result_rect.pkl", "rb") as f2:
    result_rect = pickle.load(f2)
with open("./pulse_optimisation/result_sine.pkl", "rb") as f3:
    result_sine = pickle.load(f3)
```

```python
A_drag_max, A_rect_max, A_sine_max = (0.12526195446779698, 0.031415926535897934, 0.04934802200544679)
d_min, d_max, d_num = (-0.03, 0.03, 101)
A_min, A_num = (0, 100)
d_range = np.linspace(d_min, d_max, d_num)

A_drag_range = np.linspace(A_min, A_drag_max, A_num)
A_rect_range = np.linspace(A_min, A_rect_max, A_num)
A_sine_range = np.linspace(A_min, A_sine_max, A_num)
tr_probs_drag = np.array([[np.abs(r[1,1]) for r in res] for res in result_drag])
tr_probs_rect = np.array([[np.abs(r[1,1]) for r in res] for res in result_rect])
tr_probs_sine = np.array([[np.abs(r[1,1]) for r in res] for res in result_sine])
```

```python
fig, ax = plt.subplots(1,1)
cmap = matplotlib.colormaps["cividis"]  # Choose a colormap
im = ax.pcolormesh(d_range, A_drag_range, tr_probs_drag.T, vmin=0, vmax=1, cmap=cmap)
fig.colorbar(im, ax=ax, label='Transition Probability')
plt.ylabel('Rabi Frequency (1/T)')
plt.xlabel('Detuning (1/T)')
plt.title('Transition Probability for DRAG Pulse')
plt.show()
```

```python
def leakage(result_optim):
    leakage_values = []
    for res in result_optim:
        leakage_line = []
        for r in res:
            leakage = 0
            for d in range(2, 8):
                leakage += np.abs(r[d,d])
            leakage_line.append(leakage)
        leakage_values.append(leakage_line)
    return np.array(leakage_values)
    
leakage_drag = leakage(result_drag)
leakage_rect = leakage(result_rect)
leakage_sine = leakage(result_sine)
```
